Finding all the rows of specific countries

Putting them in a dictionary



```python
countries = ['Ukraine', 'Russia', 'Germany', 'Poland', 'Australia', 'China', 'Iran', 'Brazil', 'India', 'Taiwan']
country_dfs = {}

for country in countries:
    country_dfs[country] = df[df['Country'].str.contains(country, na=False)]




```

Output the articles of each coutry in their own folder

```python
countries1 = ['Ukraine', 'Russia', 'UK', 'US', 'New Zealand']

country_dfs1 = {}

for country in countries1:
    country_dfs1[country] = df[df['Country'].str.contains(country, na=False)]
```

```python
import os


# Looping through each country and its corresponding DataFrame in the country_dfs dictionary
for country, country_df in country_dfs.items():
    #creating country folders
    country_folder = f'Country_corpus/{country}'
     # os.makedirs is used to create the directory; exist_ok=True means it won't throw an error if the folder already exists
    os.makedirs(country_folder, exist_ok=True)

    # Looping through each row in the country's DataFrame
    for index, row in country_df.iterrows():
        # Cleaning the web title to make it a valid filename. Replaces '/' and '\' with '-'
        web_title = str(row['WebTitle']).replace('/', '-').replace('\\', '-')
        # Creating the full file path by joining the country folder path and the cleaned web title
        filename = os.path.join(country_folder, f"{web_title}.txt")

        # Creating the full file path by joining the country folder path and the cleaned web title 
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(str(row['Text']))
```
