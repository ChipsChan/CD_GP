```python
import pandas as pd
```

```python
file_path = 'new_metadata_with_country_and_continent .csv'
data = pd.read_csv(file_path)
```

```python


america_rows = data[data['WebTitle'].str.contains('America|US|United States')]


america_rows
```

```python

selected_values = america_rows['Unnamed: 0'].tolist()


print(selected_values)
```

```python
import pandas as pd
import numpy as np  



new_string1 = 'US'
new_string2 = 'North America'


for row_index in selected_values:
    
    current_value1 = data.at[row_index, 'Country']
    if pd.isna(current_value1):
        data.at[row_index, 'Country'] = new_string1
    else:
        if new_string1 not in str(current_value1):
            new_value1 = f"{current_value1}, {new_string1}"
            data.at[row_index, 'Country'] = new_value1

    current_value2 = data.at[row_index, 'Continent']
    if pd.isna(current_value2):
        data.at[row_index, 'Continent'] = new_string2
    else:
        if new_string2 not in str(current_value2):
            new_value2 = f"{current_value2}, {new_string2}"
            data.at[row_index, 'Continent'] = new_value2

```
