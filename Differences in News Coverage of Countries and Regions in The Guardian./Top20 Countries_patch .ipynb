# Top 10 Countries in the Guardian world news

Based on the metadata.csv, our group aims to do further exploration about the world news section in the Guardian. We explored top 10 countries that are frequently mentioned in the world news in 2022, and try to find which country the Guardian focused most on the world news.The final result shows that in 2022, the Guardian focuses on Urkaine and Russia way more than other countries, which shows the great influence of Urkaine - Russia War.

Note: I pip install geotext in my anaconda shell prompt because geotext can extract country and city mentions from text.

```python
import pandas as pd

# Load the CSV file
file_path = 'country_uk_us_v3.csv'
data = pd.read_csv(file_path)

# Extracting the 'Country' column
countries_column = data['Country']

# Splitting the countries in each cell, accounting for cells with multiple countries separated by commas
# and cells that might be empty
all_countries = countries_column.str.split(',').dropna()
flat_list = [country.strip() for sublist in all_countries for country in sublist]

# Counting the occurrences of each country
country_counts = pd.Series(flat_list).value_counts()

# Getting the top 20 countries
top_20_countries = country_counts.head(20)
top_20_countries
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming 'top_20_countries' is already defined as in the previous code

# Setting the style for the plot
sns.set(style="whitegrid")

# Creating a custom color palette from green to blue
green_blue_palette = sns.color_palette("GnBu", len(top_20_countries))

# Creating the plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=top_20_countries.values, y=top_20_countries.index, palette=green_blue_palette)

# Adding titles and labels
plt.title('Top 20 Countries by Occurrence')
plt.xlabel('Number of Occurrences')
plt.ylabel('Country')

# Remove axes splines
for spine in bar_plot.spines.values():
    spine.set_visible(False)

# Remove background color
bar_plot.set_facecolor('none')

# Displaying the plot in the output
plt.show()

# Saving the plot with a transparent background
plt.savefig('barplot_transparent.png', bbox_inches='tight', transparent=True)


```
